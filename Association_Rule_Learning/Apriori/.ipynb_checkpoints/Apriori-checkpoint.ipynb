```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Market_Basket_Optimisation.csv',header=None)
df.head()
```

```python
df.info()
```

```python
transactions = df.astype(str).to_numpy().tolist()
```

```python
from apyori import apriori
min_sup = 3*7/7500 # Product purchased 3 times aweek
rules = apriori(transactions,min_support=min_sup,min_confidence = 0.2,min_lift=3,min_length=2)
```

```python
results = list(rules)
results
```

```python
from collections import namedtuple
RelationRecord = namedtuple('RelationRecord', ['items', 'support', 'ordered_statistics'])
OrderedStatistic = namedtuple('OrderedStatistic', ['items_base', 'items_add', 'confidence', 'lift'])

def format_relation_records(records):
    formatted_records = []
    for record in records:
        items_str = ', '.join(record.items)
        support = record.support
        for stat in record.ordered_statistics:
            base_items_str = ', '.join(stat.items_base)
            add_items_str = ', '.join(stat.items_add)
            confidence = stat.confidence
            lift = stat.lift
            
            formatted_records.append(
                f"Items: {items_str}\n"
                f"Support: {support:.4f}\n"
                f"Base Items: {base_items_str}\n"
                f"Add Items: {add_items_str}\n"
                f"Confidence: {confidence:.4f}\n"
                f"Lift: {lift:.4f}\n"
                "---------------------------"
            )
    
    return '\n'.join(formatted_records)

# Print formatted output
formatted_output = format_relation_records(results)
```
